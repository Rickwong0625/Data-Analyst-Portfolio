# Merge,Join, and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\MYSP6023\Downloads\Panda Tutorial\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\MYSP6023\Downloads\Panda Tutorial\LOTR 2.csv")
df2
```

```python
#merging DataFrames
```

```python
#inner join
df1.merge(df2,how = 'inner',on=['FellowshipID','FirstName'])
```

```python
#outer join
df1.merge(df2,how = 'outer')
```

```python
#left join
df1.merge(df2,how = 'left')
```

```python
#right join
df1.merge(df2,how = 'right')
```

```python
#cross join
df1.merge(df2,how = 'cross')
```

```python
#join function
```

```python
df1.join(df2,on = 'FellowshipID',how ='outer',lsuffix='_left',rsuffix = 'right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_left',rsuffix = 'right',how='outer')
df4
```

```python
#concat function
pd.concat([df1,df2],join = 'outer',axis=1)
```

```python
#append function
df1.append(df2)
```
